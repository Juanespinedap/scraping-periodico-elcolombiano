```python
# Librerias

import requests
from bs4 import BeautifulSoup
from IPython.display import Image
```

```python
pagina_inicial = requests.get("https://www.elcolombiano.com")
pagina_inicial = BeautifulSoup(pagina_inicial.text, "html.parser")
imagen_inicial = pagina_inicial.find(name="section", attrs={"class" : "ec-teaser-noticia-principal"}).find("img")
imagen_url = str("https://" + imagen_inicial.get("src")[2:])
imagen_url
imagen = requests.get(imagen_url)
Image(imagen.content)
```
